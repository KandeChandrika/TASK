```python
# 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot
df = pd.read_csv(r"C:\Users\chand\Data science\df_stats_cleaned.csv")
```

```python
df.groupby(["fuelType","gearbox"])["price"].mean().plot(kind = "bar",color = ["green","orange"])
```

```python
pd.crosstab(df['fuelType'],df['gearbox'], values = df['price'], aggfunc = 'mean').plot(kind = 'bar', xlabel = "fuel type", ylabel = 'Average Price of vehicle')
```

# 3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
pd.crosstab(df['vehicleType'],df['gearbox'], values = df['powerPS'], aggfunc = 'mean')
```

```python
pd.crosstab(df['vehicleType'],df['gearbox'], values = df['powerPS'], aggfunc = 'mean').plot(kind = 'bar',ylabel = 'avg power')
```
